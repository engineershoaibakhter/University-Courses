# **Bisection Method**

```python
def function(x):
  return x**2-4


def bisection_method(a,b,num_iteration):
  if(function(a)*function(b)>=0):
    print('incorrect intervals')
    return
  for _ in range(num_iteration):
    c=(a+b)/2

    if function(c)==0:
      print('exact root= ',c)
      return

    if(function(c)*function(a)<0):
      b=c
    else:
      a=c

  print('approximate root: ',c )
```

```python
a=1
b=3
iterations=10
bisection_method(a,b,iterations)
```

# **Regular Falsy**

```python
def function(x):
  return (((x-4)**2)*(x+2))


def bisection_method(a,b,num_iteration):
  if(function(a)*function(b)>=0):
    print('incorrect intervals')
    return
  for _ in range(num_iteration):
    c=a-((function(a)*(b-a))/(function(b)-function(a)))

    if function(c)==0:
      print('exact root= ',c)
      return

    if(function(c)*function(a)<0):
      b=c
    else:
      a=c

  print('approximate root: ',c )
```

```python
a=-2.5
b=-1.0
iterations=3
bisection_method(a,b,iterations)
```
